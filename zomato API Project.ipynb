### In this question we have to plot the pie graph of top 10 cuisines present in restaurants in the USA.

```python
#question 3(5)
plt.figure(figsize=(10,3))
grouped_df=file_data.groupby("City")
city_list=[]
weighted_list=[]
no_rest_list=[]

for i,j in grouped_df:
    total_votes=j["Votes"].sum()
    weighted=(j["Votes"]*j["Aggregate rating"]).sum()
    weighted_rating=weighted/total_votes
    city_list.append(i)
    weighted_list.append(weighted_rating)
    no_rest_list.append(len(j))
plt.scatter(city_list,no_rest_list,s=weighted_list*100)
plt.xticks(rotation=90)
plt.show()
```
